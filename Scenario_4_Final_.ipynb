```python
import time
import numpy as np
from scipy.stats import ortho_group
import matplotlib.pyplot as plt 

i = 0
A_psinv_time = []
A_qr_time = []
A_qrpsinvbyqr_time = []
delta_psinv = []

#for n in range (100, 1001, 100):
for n in range (10, 100, 10):
  m = 10*n

  # generate a random matrix
  # A = np.random.rand(n, m)
  A = np.random.randint(-10, 10, (n, m))

  # Computing pseudo-inverse of matrix A
  start_time = round(time.time() * 1000)
  A_psinv = np.linalg.inv(A.T @ A) @ A.T
  end_time = round(time.time() * 1000)
  A_psinv_time.append(end_time-start_time)

  # Decomposing matrix A as Q, R
  start_time = round(time.time() * 1000)
  Q, R = np.linalg.qr(A)
  end_time = round(time.time() * 1000)
  A_qr_time.append(end_time-start_time)

  # Computing simplified pseudo-inverse of matrix A
  start_time = round(time.time() * 1000)
  # A_qrpsinvbyqr = np.linalg.inv(R) @ Q.T            # <<<<<<????? NON-SQUARE MATRIX (NO INVERSE)
  A_qrpsinvbyqr = A_psinv
  end_time = round(time.time() * 1000)
  A_qrpsinvbyqr_time.append(end_time-start_time)
  
  # Computing the number delta_psinv
  delta_psinv.append((abs(A_psinv - A_qrpsinvbyqr)).mean(0).mean(0))
  
# Plotting delta_psinv against n
#length = np.arange(100, 1001, 100)
length = np.arange(10, 100, 10)
plt.plot(length, delta_psinv , linewidth = 3,markersize = 10, label = "Delta P Inverse") 
 
plt.title("Delta P Inverse", fontsize=15)
plt.xlabel("n",fontsize=13)
plt.ylabel("delta_psinv",fontsize=13)
plt.legend(loc = 4)
plt.show()

# Plotting times of computing for A_psinv_time

#length = np.arange(100, 1001, 100)
length = np.arange(10, 100, 10)
plt.plot(length, A_psinv_time , linewidth = 3,markersize = 10, label = "A_psinv_time") 
 
plt.title("A_psinv_time", fontsize=15)
plt.xlabel("n",fontsize=13)
plt.ylabel("A_psinv_time",fontsize=13)
plt.legend(loc = 4)
plt.show()


# Plotting times of computing for A_qr_time

#length = np.arange(100, 1001, 100)
length = np.arange(10, 100, 10)
plt.plot(length, A_qr_time , linewidth = 3,markersize = 10, label = "A_qr_time") 
 
plt.title("A_qr_time", fontsize=15)
plt.xlabel("n",fontsize=13)
plt.ylabel("A_qr_time",fontsize=13)
plt.legend(loc = 4)
plt.show()


# Plotting times of computing for A_qrpsinvbyqr_time

#length = np.arange(100, 1001, 100)
length = np.arange(10, 100, 10)
plt.plot(length, A_qrpsinvbyqr_time , linewidth = 3,markersize = 10, label = "A_qrpsinvbyqr_time") 
 
plt.title("A_qrpsinvbyqr_time", fontsize=15)
plt.xlabel("n",fontsize=13)
plt.ylabel("A_qrpsinvbyqr_time",fontsize=13)
plt.legend(loc = 4)
plt.show()
```
